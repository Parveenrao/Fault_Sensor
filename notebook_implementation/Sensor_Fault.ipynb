### Columns With Zero Standard Deviation And Missing value > 70%

```python
def get_cols_with_std(df : pd.DataFrame):
    cols_to_drop = []
    numerical_columns = [col for col in sensor.columns if  sensor[col].dtype != 'O']
    for col in  numerical_columns:
        if sensor[col].std() == 0:
            cols_to_drop.append(col)

    return cols_to_drop

    
    
```

```python
def get_redundant_cols(df: pd.DataFrame, missing_threshold = 0.7):

    cols_missing_ratios = sensor.isna().sum().div(sensor.shape[0])
    cols_to_drop = list(cols_missing_ratios[cols_missing_ratios > missing_threshold].index)
    return cols_to_drop    
```

```python
cols_to_drop1 = get_redundant_cols(sensor , missing_threshold = 0.7)
```

```python
cols_to_drop1
```

```python
cols_to_drop2 = get_cols_with_std(df = sensor)
```

```python
cols_to_drop2
```

```python
cols_to_drop = cols_to_drop1 + cols_to_drop2
```

```python
cols_to_drop
```

```python
sensor = sensor.drop(cols_to_drop , axis = 1)
```

## Data Transformation

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.impute import SimpleImputer
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer(n_neighbors =3)
preprocessing_pipeline = Pipeline(steps =[
    ('Imputer' , imputer),
    ('Scaler', RobustScaler())
])
```

```python
preprocessing_pipeline 
```

```python
X = X.drop('Unnamed: 0' , axis =1)
```

```python
X_transformed = preprocessing_pipeline.fit_transform(X)
```

```python
X_transformed
```

## Resampling

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import TomekLinks
```

```python

X, y = X_transformed[:, :-1], y

try:
    oversampler = SMOTE(sampling_strategy="auto", k_neighbors=4)
    undersampler = TomekLinks()
    X_resampled, y_resampled = oversampler.fit_resample(X, y)
    X_res, y_res = undersampler.fit_resample(X_resampled, y_resampled)
except ValueError as e:
    print("An error occurred during resampling: ", str(e))
```

```python
X_res.shape
```

```python
y_res
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)

print(f"train set: {X_train.shape, y_train.shape}")
print(f"test set: {X_test.shape, y_test.shape}")
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import roc_auc_score

# Shortlisted base Models
svc_clf = SVC(kernel='linear')
svc_rbf_clf = SVC(kernel='rbf')
random_clf = RandomForestClassifier(random_state=42)
xgb_clf = XGBClassifier(objective='binary:logistic')
```

```python
svc_model = svc_clf.fit(X_train  ,y_train)
svc_model
```

```python
y_pred = svc_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", accuracy)
```
